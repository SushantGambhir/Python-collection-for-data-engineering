# Joining Data in Pandas Dataframes

Here are the steps involved in joining data using Pandas Dataframe APIs.

- First, we need to define index for both the dataframes that are supposed to be joined.
- Use join to join the Dataframes based on index. The default join is left, we need to make sure right join is used based on the requirement.

In our case, we will join orders and customers using inner join.

```python
import json
```

```python
import pandas as pd
```

```python
# Define a helper function that returns sorted column names for a given table schema.
def get_column_names(schemas, table_name):
    column_details = schemas[table_name]
    columns = sorted(column_details, key = lambda col: col['column_position'])
    return [col['column_name'] for col in columns]
```

```python
schemas = json.load(open('data/retail_db/schemas.json'))
```

```python
orders_columns = get_column_names(schemas,'orders')
```

```python
orders = pd.read_csv('data/retail_db/orders/part-00000',names=orders_columns)
```

```python
# Now we need to perform join of 'orders' and 'customers'
# The column used for join: order_customer_id (orders) & customer_id (customers)
```

```python
# Now lets join both the dataframes

customer_orders = customers.join(orders, how='inner')
```

The statement `customer_orders = customers.join(orders, how='inner')` performs an inner join between the `customers` and `orders` DataFrames in Pandas, based on their indices. Here's a breakdown of what it does:

### Key Concepts
- **Join in Pandas**: The `join` method combines two DataFrames by matching rows based on their indices (not columns, unless specified). It's similar to SQL joins but operates on DataFrame indices by default.
- **Indices in this context**:
  - The `customers` DataFrame has its index set to `'customer_id'` (from an earlier cell: `customers = customers.set_index('customer_id')`).
  - The `orders` DataFrame has its index set to `'order_customer_id'` (from an earlier cell: `orders = orders.set_index('order_customer_id')`).
  - Since `'order_customer_id'` in `orders` corresponds to `'customer_id'` in `customers` (based on the data schema), the indices represent the same logical key for matching customers to their orders.
- **Inner join (`how='inner'`)**: This keeps only the rows where the index values match in both DataFrames. Rows without a match in either DataFrame are excluded.

### What the Statement Does
1. **Matching criteria**: It matches rows where the index of `customers` (i.e., `customer_id` values) equals the index of `orders` (i.e., `order_customer_id` values).
   - For example, if a customer with `customer_id = 1` has orders, those rows will be combined.
   - If a customer has no orders or an order has no matching customer, those rows are dropped.

2. **Result structure**:
   - The resulting `customer_orders` DataFrame includes:
     - All columns from `customers` (e.g., `customer_fname`, `customer_lname`, etc.).
     - All columns from `orders` (e.g., `order_id`, `order_date`, `order_status`).
   - The index of the result is the index from the left DataFrame (`customers`), which is `customer_id`.
   - Duplicate column names (if any) are handled by Pandas (e.g., suffixes like `_x` and `_y` are added, but in this case, the columns are distinct).

3. **Example behavior**:
   - Suppose `customers` has a row for `customer_id = 1` (e.g., Richard Hernandez).
   - And `orders` has rows where `order_customer_id = 1` (e.g., multiple orders for that customer).
   - The join will create combined rows for each matching order, duplicating the customer info for each order.

4. **Why inner join?**: 
   - It ensures only customers with orders (and orders with valid customers) are included.
   - This is appropriate here since the goal is to analyze customer-order relationships, excluding unmatched data.

### Potential Outcomes
- If there are multiple orders per customer, the result will have multiple rows per customer (one per order).
- The shape of `customer_orders` will be smaller than or equal to the combined shapes of `customers` and `orders`, depending on matches.
- From the notebook, you can check shapes: `customers.shape`, `orders.shape`, and `customer_orders.shape` to see the impact.

```python
customer_orders.reset_index()
```

```python
customer_orders.reset_index().groupby('customer_id')['customer_id'].agg(order_count='count').reset_index()
```

This Pandas expression counts how many orders each customer has in the `customer_orders` DataFrame.

### Step-by-step

- `customer_orders.reset_index()`
  - Converts the index back into a regular column.
  - In this file, the index is `customer_id`, so after this step `customer_id` becomes a normal column again.

- `.groupby('customer_id')`
  - Groups the rows by `customer_id`.
  - Each group now represents all orders for a single customer.

- `['customer_id']`
  - Selects the `customer_id` column from the grouped data.
  - This is the column used for the aggregation.

- `.agg(order_count='count')`
  - Aggregates each group by counting the number of `customer_id` values.
  - Since each row is one order, this count is effectively the number of orders per customer.
  - The result is a new column named `order_count`.

- `.reset_index()`
  - Turns the grouped index (`customer_id`) back into a normal column.
  - This gives a DataFrame with two columns: `customer_id` and `order_count`.

### Final result

A DataFrame showing:
- `customer_id`
- `order_count`

So it answers: “How many orders does each customer have?”
