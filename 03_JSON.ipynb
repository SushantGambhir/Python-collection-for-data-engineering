```python
# Set the path to the JSON file that contains table schema definitions.
file_path = 'data/retail_db/schemas.json'
```

```python
# Load the JSON file into a Python dictionary containing schema definitions.
schemas = json.load(open(file_path,'r'))
```

```python
# Verify the loaded schema data is a Python dictionary.
type(schemas)
```

```python
# Retrieve the schema details for the 'orders' table using dictionary lookup.
schemas.get('orders')
```

```python
# Access the 'orders' schema directly by key to inspect its column definitions.
schemas['orders']
```

```python
# Access the schema for the 'categories' table to compare structure across tables.
schemas['categories']
```

```python
# List all table names available in the schema JSON file.
schemas.keys()
```

```python
# Access the schema for the 'departments' table to inspect its column metadata.
schemas['departments']
```

```python
# Store the orders table schema in a variable and display the list of column definitions.
column_details = schemas['orders']
column_details
```

```python
# Show the first column definition from the orders schema list.
column_details[0]
```

```python
# Getting all column names from the orders table schema using list comprehension.
[col['column_name'] for col in column_details]
```

```python
# Getting all column names from the orders table schema using map.
list(map(lambda col: col['column_name'], column_details))
```

```python
# Display the full list of order table column metadata again for review.
column_details
```

```python
# Ask the user which schema field should be used as the sort key.
sorting_key = input("Enter the column name to sort by: ")
print("Sorting by column: "+ sorting_key)
```

```python
# Sort the order table schema list by the user-selected field, such as column_position or column_name.
columns = sorted(column_details, key=lambda col: col[sorting_key])
```

```python
# Extract just the column names from the sorted schema list for easy viewing.
[col['column_name'] for col in columns]
```

```python
# Define a helper function that returns sorted column names for a given table schema.
def get_column_names(schemas, table_name,sorting_key):
    column_details = schemas[table_name]
    columns = sorted(column_details, key = lambda col: col[sorting_key])
    return [col['column_name'] for col in columns]
```

```python
# Call the helper function to get the orders table column names sorted by position.
get_column_names(schemas,'orders','column_position')
```
