```python
import pandas as pd
```

```python
pd.read_csv('data/retail_db/orders/part-00000')

# Will create a dataframe. Since column names are not defined, it will take first row as header
```

```python
pd.read_csv('data/retail_db/orders/part-00000',header=None)

# Now it wont consider first row as header
```

```python
orders_columns = ['order_id', 'order_date','order_customer_id', 'order_status']
```

```python
pd.read_csv('data/retail_db/orders/part-00000',names=orders_columns)

# Will have appropriate column names
```

```python
orders = pd.read_csv('data/retail_db/orders/part-00000',names=orders_columns)
```

```python
orders
```

```python
orders.columns

# Returns all the column names in the orders dataframe
```

```python
orders['order_status'].unique()

# Returns all the unique values in order_status column
```

```python
orders.query("order_status == 'COMPLETE'")

# Returns all the entries where order_status is COMPLETE
```

```python
orders.query("order_status == 'COMPLETE' and order_date == '2014-07-06 00:00:00.0'")

# Returns entries with COMPLETE status and specified order date
```

```python
orders.query("order_status == 'COMPLETE' or order_status == 'CLOSED'")

# Returns entries with either COMPLETE or CLOSED status

# This statement can also be written as:
# orders.query("order_status == ('COMPLETE','CLOSED')")
```
